```python
features = ['cap-shape', 'cap-surface', 'cap-color', 'bruises', 'odor', 'gill-attachment', 
            'gill-spacing', 'gill-size', 'gill-color', 'stalk-shape', 'stalk-root', 
            'stalk-surface-above-ring', 'stalk-surface-below-ring', 'stalk-color-above-ring', 
            'stalk-color-below-ring', 'veil-type', 'veil-color', 'ring-number', 
            'ring-type', 'spore-print-color', 'population', 'habitat']

palette = {'e':'#34d14c', 'p': '#b50505'}
hueOrder=['e','p']

for feature in features:
    #plt.figure(figsize=(12, 6))
    sns.countplot(data=df, x=feature, hue='class', palette=palette, hue_order=hueOrder)
    sns.set(rc={'axes.facecolor':'#f5f3f1', 'figure.facecolor':'#f5f3f1', 'axes.edgecolor':'grey', 'axes.grid':True, 'grid.color':'white'})
    plt.title(f'Comparison of {feature.capitalize()} between Poisonous and Edible Mushrooms')
    plt.xlabel(feature)
    plt.legend(title='Key', labels=['Edible', 'Poisonous'])
    plt.savefig(f'visuals/{feature}Comparison.png')
    plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='spore-print-color', y='bruises', hue='class', data=df, split=True, palette='Pastel1')
plt.title('Violin Plot of Gill Spacing by Cap Shape and Class')
plt.show()
```

```python
sns.pairplot(data = df, vars='class', hue='class')#, palette={'Edible': 'green', 'Poisonous': 'red'})
```

```python
sections = {
    'Cap': ['cap-shape', 'cap-surface', 'cap-color', 'bruises'],
    'Gill': ['gill-attachment', 'gill-spacing', 'gill-size', 'gill-color'],
    'Stalk': ['stalk-shape', 'stalk-root', 'stalk-surface-above-ring', 
              'stalk-surface-below-ring', 'stalk-color-above-ring', 'stalk-color-below-ring'],
    'Veil': ['veil-type', 'veil-color'],
    'Ring': ['ring-number', 'ring-type']
}

for section, features in sections.items():
    plt.figure(figsize=(15, 10))
    plt.suptitle(f'{section} Features Comparison Between Poisonous and Edible Mushrooms', fontsize=16)

    for i, feature in enumerate(features):
        plt.subplot(2, 3, i + 1)
        sns.countplot(data=df, x=feature, hue='class')
        sns.set(rc={'axes.facecolor':'#f5f3f1', 'figure.facecolor':'#f5f3f1', 'axes.edgecolor':'grey', 'axes.grid':True, 'grid.color':'white'})
        plt.title(f'{feature.capitalize()}')
        plt.xlabel(feature.capitalize())
        plt.ylabel('Count')
        plt.legend(loc='upper right')

    plt.tight_layout(rect=[0, 0, 1, 0.95])
    plt.show()
```

```python
for section, features in sections.items():
    plt.figure(figsize=(15, 10))
    plt.suptitle(f'{section} Features Comparison Between Poisonous and Edible Mushrooms', fontsize=16)
    
    for i, feature in enumerate(features):
        plt.subplot(2, 3, i + 1)
        cross_tab = pd.crosstab(df[feature], df['class'], normalize='index')
        sns.heatmap(cross_tab, annot=True, cmap='coolwarm', ax=plt.gca())
        plt.title(f'{feature.capitalize()}')
        plt.xlabel('Class')
        plt.ylabel(feature.capitalize())
    
    plt.tight_layout(rect=[0, 0, 1, 0.95])
    plt.savefig(f'visuals/{feature}HeatMap.png')
    plt.show()
    
```
